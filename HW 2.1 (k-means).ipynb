```python
import pandas as pd

df = pd.read_csv('mobile.csv', index_col=0)
df.head()
```

```python
chosen_data = df[['battery_power', 'clock_speed', 'int_memory', 'ram']]
chosen_data.head()
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=5)
kmeans.fit(chosen_data)
kmeans.inertia_
```
